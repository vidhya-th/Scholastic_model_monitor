## Student Performance Indicator


#### Life cycle of Machine learning Project

- Problem Statement: Identifying the objective, success metrics, and type of learning (Regression/Classification).

- Data Ingestion: Extracting raw data from sources.

- Validation & Quality Assurance: Checking for null values, duplicates, and data type consistency.

- Exploratory Data Analysis (EDA): Investigating feature distributions, multicollinearity, and statistical correlations.

- Feature Engineering & Pre-processing: Performing encoding, normalization, and handling missing data.

- Model Development & Training: Partitioning data into train/test sets and fitting algorithms.

- Model Evaluation & Hyperparameter Optimization: Assessing performance via cross-validation and tuning for the optimal configuration.

### 1) Problem statement
- This project aims to quantify how external factors—such as demographics (Gender, Ethnicity) and socio-economics (Parental Education, Lunch programs)—influence a student's academic output.

- Technically, the "Problem Statement" involves identifying the dependent variable (test scores) and analyzing its relationship with multiple independent variables. By establishing these links, we can determine which features have the highest predictive power or "feature importance."

- This project is a Regression problem.The target variables (test scores in Math, Reading, and Writing) are continuous numerical values.

### 3.8 Adding columns for "Total Score" and "Average"

```python
df['total score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total score']/3
df.head()
```

```python
reading_full = df[df['reading_score'] == 100]['average'].count()
writing_full = df[df['writing_score'] == 100]['average'].count()
math_full = df[df['math_score'] == 100]['average'].count()

print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
```

```python
reading_less_20 = df[df['reading_score'] <= 20]['average'].count()
writing_less_20 = df[df['writing_score'] <= 20]['average'].count()
math_less_20 = df[df['math_score'] <= 20]['average'].count()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```

### 5. Conclusions
- Socio-Economic & Demographic Correlation: Academic performance shows a statistically significant relationship with lunch type (socio-economic status), race/ethnicity, and parental level of education.
